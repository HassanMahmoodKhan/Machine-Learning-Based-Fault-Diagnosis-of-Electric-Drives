```python
new_df = new_df.drop(['time (sec)','Torque (N*m)'], axis=1)
new_df
```

```python
x = new_df.drop(['Category'], axis=1)
x.info()
```

```python
y = new_df['Category']
y.info()
```

```python
numeric_transformer = make_pipeline(
    SimpleImputer(strategy = "mean"),
    MinMaxScaler()
)
numeric_transformer.fit(x)
```

```python
X = numeric_transformer.transform(x)
X = pd.DataFrame(X, columns = x.columns)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2) 
# splitting the dataset into training and testing subset
```

```python
cv = KFold(n_splits=5)
params = [{'n_neighbors': np.arange(1, 50, 2),
           'weights': ['uniform', 'distance'],
           'p': [1,2],
           'n_jobs': [-1]
          }]
knn_clf = KNeighborsClassifier()
clf = GridSearchCV(knn_clf,
                      param_grid=params,
                      scoring='accuracy',
                      cv=cv)
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=23, weights='uniform', p=1, n_jobs=-1 )
cv = KFold(n_splits=10)
scores_knn = pd.DataFrame(cross_validate(knn_clf, X_train, y_train, scoring= 'accuracy', cv=cv, n_jobs=-1, 
error_score='raise', return_train_score=True))
```

```python
scores_knn.mean()
```

```python
knn_clf.fit(X_train, y_train)
knn_pred=knn_clf.predict(X_test)
accuracy_score(y_test, knn_pred)
```

```python
cm = confusion_matrix(y_test, knn_pred, normalize='all')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["NOM", "PTPF","PTGF", "OVF", "OLF", "UVF"])
sns.set_style("white")
plt.rc('font', size=10)
disp.plot()
plt.show()
```

```python
cv = KFold(n_splits=5)
params = [{'n_estimators': np.arange(start = 5, stop = 105, step = 5),
         'criterion': ['gini'],
         'max_depth': np.arange(start = 1, stop = 20, step = 1),
         'max_features':['sqrt', 'log2'],
         'min_samples_leaf': np.arange(start = 1, stop = 10, step = 1),
           'n_jobs': [-1]
          }]
rf_clf = RandomForestClassifier()
clf = GridSearchCV(rf_clf,
                      param_grid=params,
                      scoring='accuracy',
                      cv=cv)
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
rf_clf = RandomForestClassifier(criterion = 'gini', max_depth = 10, max_features = 'sqrt', min_samples_leaf = 9,
                                n_estimators = 50, n_jobs = -1 )
cv = KFold(n_splits=10)
scores_rf = pd.DataFrame(cross_validate(rf_clf, X_train, y_train, scoring= 'accuracy', cv=cv, n_jobs=-1, 
error_score='raise', return_train_score=True))
scores_rf.mean()
```

```python
rf_clf.fit(X_train,y_train)
rf_pred = rf_clf.predict(X_test)
accuracy_score(y_test, rf_pred)
```

```python
cm = confusion_matrix(y_test, rf_pred, normalize='all')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["NOM", "PTPF","PTGF", "OVF", "OLF", "UVF"])
sns.set_style("white")
plt.rc('font', size=10)
disp.plot()
plt.show()
```
